```python
x=df[['Away Team'	,'Away Score'	,'Home Team'	,'Home Score']]
y=df['Winner']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy as np

lb=LabelEncoder()
y_enco=lb.fit_transform(y)
cat=['Away Team', 'Home Team']
x_enco=pd.get_dummies(x,columns=cat,drop_first=True)
x_train,x_test,y_train_enco,y_test_enco=train_test_split(x_enco,y_enco,test_size=0.2,random_state=20 )
```

```python
y_train_enco
```

```python
x_test.head()
```

```python
from xgboost import XGBClassifier

bst=XGBClassifier(n_estimators=150, max_depth=7,objective='binary:logistic',random_state=20)
bst.fit(x_train,y_train_enco)
pred=bst.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test_enco,pred)
print(f"Accuracy:{accuracy}")
```

```python
pred_encoded = bst.predict(x_test)

# Convert the predicted encoded labels back to the original class names (team names)
pred_teams = lb.inverse_transform(pred_encoded)
actual_teams = lb.inverse_transform(y_test_enco)
results_df = pd.DataFrame({'Actual Winner': actual_teams, 'Predicted Winner': pred_teams})
print(results_df.sample(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Assuming you have the true and predicted team names
# Example:
pred_encoded = bst.predict(x_test)

# Convert the predicted encoded labels back to the original class names (team names)
pred_teams = lb.inverse_transform(pred_encoded)
actual_teams = lb.inverse_transform(y_test_enco)

actual_counts = Counter(actual_teams)
pred_counts = Counter(pred_teams)

df_counts = pd.DataFrame({
    'Actual': pd.Series(actual_counts),
    'Predicted': pd.Series(pred_counts)
}).fillna(0).astype(int)

df_counts.plot(kind='bar', figsize=(12,6))
plt.title('Actual vs Predicted Wins per Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
